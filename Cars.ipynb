### Outlier detection in the dataset

Sometimes a dataset can contain extreme values that are outside the range of what is expected and unlike the other data. These are called outliers and often machine learning modeling and model skill in general can be improved by understanding and even removing these outlier values.

In machine learning, outliers can have a significant impact on the performance of a model, as they can skew the training      process and lead to incorrect predictions.

Therefore, it is important to detect and remove outliers from the data before building a machine learning model.
There are many techiniques to detect outliers. Let us first see the simplest form of visualizing outliers.

Box plots are a graphical depiction of numerical data through their quantiles. It is a very simple but effective way to visualize outliers. 
Think about the lower and upper whiskers as the boundaries of the data distribution. Any data points that show above or below the whiskers, can be considered outliers or anomalous.

```python
#plotting boxplots for outlier detection
for column in num_cols:
    sns.boxplot(x =num_cols[column])
    plt.show()
```

#### Observation:
Here boxplots show the proper distribution of of 25 percentile and 75 percentile of numerical features of the dataset.

### Outlier Removal Techniques

#### Using IQR technique

Inter Quartile Range (IQR) is one of the most extensively used procedure for outlier detection and removal.

According to this procedure, we need to follow the following steps:

Find the first quartile, Q1.

Find the third quartile, Q3.

Calculate the IQR. IQR = Q3-Q1.

Define the normal data range with lower limit as Q1–1.5*IQR and upper limit as Q3+1.5*IQR.

Any data point outside this range is considered as outlier and should be treated or removed for further analysis.

In boxplot, this IQR method is implemented to detect any extreme data point where the maximum point (the end of high whisker) is Q3+1.5*IQR and the minimum point (the start of low whisker) is Q1–1.5*IQR.

```python
# outlier detection using IQR
for col in num_cols:
 Q1 = df[col].quantile(0.25)
 Q3 = df[col].quantile(0.75)
 IQR = Q3-Q1
 lower_limit = Q1 - 1.5*IQR
 upper_limit = Q3 + 1.5*IQR
 print(df[(df[col]>upper_limit) | (df[col]<lower_limit)])
```

#### Note :
By the way there are many other techniques by which we can detect outliers such as Z-score methods , by using clustering algorithms etc.

Outlier removal is a critical preprocessing step in machine learning, but it should be performed with caution.
Domain knowledge is important in better understanding outliers and continuing forward with the necessary steps to treat them.

Completely removing too many outliers can result in the loss of important information and may lead to underfitting.
On the other hand if you try to replace them with mean, median or something else you can  create a bias in the data which may affect the model's accuracy.

Therefore it is important to carefully analyze the data and use a combination of methods to identify and treat outliers.
