## Analysis and Visualizations

### 1. Which Category is Best Selling and Most Profitable?

We will group the Sales, Profit and also Quantity by Category.

```python
# grouping Sales, profit and quantity by category
category_analysis = pd.DataFrame(superstore.groupby(['Category'])[['Sales','Profit','Quantity']].sum())
category_analysis
```

```python
# setting subplots
figure, axis = plt.subplots(1, 3, figsize=(20,10))

#plotting barcharts
category1 = sns.barplot(x =category_analysis.index,y =category_analysis.Sales,ax =axis[0])
category2 = sns.barplot(x =category_analysis.index,y =category_analysis.Profit,ax =axis[1])
category3 = sns.barplot(x =category_analysis.index,y =category_analysis.Quantity,ax =axis[2])

# Set titles
category1.set(title = 'Sales')
category2.set(title = 'Profit')
category3.set(title = 'Quantity')

# theme of the plot
sns.set_theme(style="darkgrid")

plt.show()
```

#### Our observations are:-

1- All 3 categories - Furniture, Office Supplies and Technology make similar amount of sales.

2- Technology is Best Selling and it's good to know that this category is also the Most Profitable category too. 
  Though minimal   quantity is sold in this category as these products are usually one-time purchases and can last for a long     time.
    
3- Although Furniture makes similar sales as Technology, it is the least profitable and quantity sold are at a minimum too.

4- Office Supplies sells the most in terms of quantity as it is relatively cheap product and are frequently purchased.

### 2 - Which is the most selling Sub-Category?

```python
# grouping quantity by sub-category
subcat_quantity = pd.DataFrame(superstore.groupby(['Sub-Category'])[['Quantity']].sum().sort_values('Quantity',ascending=False))
subcat_quantity
```

```python
# ploting the sub-categories
sns.set_theme(style ='darkgrid')
sns.barplot(data = subcat_quantity, x =subcat_quantity.Quantity, y= subcat_quantity.index,palette ='tab10')
plt.title('Top Selling SubCategory')
plt.show()
```

#### Our Observations are: -

As you can see the most items are being sold in the categories of Binders , Papers , Furnishings and Phones as these are frequently used and exhausted categories in various settings.

So the management should ensure that the inventory is always well-stocked for these categories.

Despite being most profitable, Copiers sell the least only 234, but as it is a relatively expensive office equipment that is usually used for few years, it is understandable that it sells the least among all.

### 3. What are the Best Selling and Most Profitable Sub-Category?

```python
# group by sales and profit sub-category
subcat_analysis = pd.DataFrame(superstore.groupby('Sub-Category')[['Sales','Profit']].sum())
subcat_analysis
```

```python
# sorting by sales(descending)
subcat_sales =pd.DataFrame(subcat_analysis.sort_values('Sales',ascending =False))
subcat_sales
```

```python
# sorting by profit
subcat_profit = pd.DataFrame(subcat_analysis.sort_values('Profit',ascending = False))
subcat_profit
```

```python
figure,axis= plt.subplots(1,2,figsize = (18,12))

# Plot Bar Plot for Best Selling Sub-Category
subcat1 = sns.barplot(data = subcat_sales, y = subcat_sales.index, x= subcat_sales.Sales, ax=axis[0] ,palette='rainbow')
subcat1.set_title('Best Selling Sub-Category')


# Plot Bar Plot for Most Profitable Sub-Category
subcat2 = sns.barplot(data = subcat_profit, y = subcat_profit.index, x = subcat_profit.Profit, ax=axis[1],palette='husl')
subcat2.set_title('Most Profitable Sub-Category')


plt.tight_layout()
plt.show()
```

#### Our Observations are:-

Phones and Chairs are the best selling sub-category.

Copiers produces most profit as they are pretty expensive followed by Phones, Accessories and  Papers . The marketing strategy is to focus on marketing these products.

On the other end of the spectrum, Machines, Fasteners, Supplies, Bookcases and Tables make close to zero margin to losses.
These are products that Super Store can consider dropping from the product catalogue or increase the sale price and profit margin or bargain for a lower price from the supplier.

### 4. Which Customer Segment is Most Profitable?

```python
# grouping by profit and sales on segments
segment_analysis = pd.DataFrame(superstore.groupby('Segment')[['Profit','Sales']].sum())
segment_analysis
```

```python
# plotting the segments
seg1 = sns.barplot(data = segment_analysis,x =segment_analysis.Profit,y =segment_analysis.index,palette ='tab10')

plt.tight_layout()
plt.show()
```

#### Our Observations are:-

Consumer segment is most profitable, followed by Corporate Segment and Home Office. Hence, marketing strategy has to target or place more focus to retain more customers from Consumer and Corporate segments.

### 5. Which is the Preferred Ship Mode?

```python
# plotting shipment modes
sns.countplot(superstore['Ship Mode'],palette ='rocket')
plt.show()
```

#### Our Observations are:-

As you can see that Standard Class is the preferred method of shipment by a long shot and perhaps the cheapest one too. The other modes are not much prefered among the customers as they  may be too expensive.

### 6. Which Region is the Most Profitable?

```python
# grouping by region
regional_analysis = pd.DataFrame(superstore.groupby('Region')['Profit'].sum().reset_index())
regional_analysis
```

```python
# plotting pie chart for regional sales and profit
explode = [ 0.1,0,0,0.2]

palette_color = sns.color_palette('coolwarm')

plt.pie(regional_analysis.Profit,labels = regional_analysis.Region ,startangle = 90,explode=explode, autopct='%.0f%%',shadow ='True',colors =palette_color)

plt.title('Most Profitable Region')
plt.show()
```

#### Our Observations:-

As seen by the pie chart the eastern and the western regions are definitely the most profitable.

### 7. Which City has the Highest Number of Sales?

```python
# grouping by  sales , profit and quantity on city
city_sales = pd.DataFrame(superstore.groupby(['City'])['Sales', 'Profit','Quantity'].sum().sort_values('Sales',ascending = False))

top_sales = city_sales[:10]
top_sales
```

```python
city_profit = pd.DataFrame(superstore.groupby(['City'])['Sales', 'Profit','Quantity'].sum().sort_values('Profit',ascending = False))

top_profit= city_profit[:10]
top_profit
```

```python
figure, axis = plt.subplots(1, 2, figsize=(18, 6))

# plotting the top selling cities
top_sales1= sns.barplot(data = top_sales, y = top_sales.index, x = top_sales.Sales, palette = "rainbow", ax = axis[0])
top_sales1.set_title('Top 10 cities with Highest Sales')

# plotting the most profitable cities
top_profit1 = sns.barplot(data = top_profit, y = top_profit.index, x = top_profit.Profit, palette = "coolwarm", ax=axis[1])
top_profit1.set_title('Top 10 most Profitable cities')

plt.tight_layout()
plt.show()
```

For loss-making products like Supplies, Bookcases, Tables, consider to either drop these from the catalogue or change suppliers and bargain for cheaper price.

Selling Bookcases and Tables result in huge losses, so Super Store has to consider to bundle them together with High Selling or Profitable sub-category such as Chairs, Copiers, Phones and Office Supplies products.
